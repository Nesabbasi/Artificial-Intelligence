```python
normalizedTrainData = (trainData - trainData.mean(numeric_only=True)) / trainData.std(numeric_only=True)
normalizedTrainData['Survived'] = trainData['Survived']
normalizedTrainData.hist(figsize=(15, 10))
```

# 11.

```python
meanSurvived = normalizedTrainData[normalizedTrainData['Survived']==1].mean()
stdSurvived = normalizedTrainData[normalizedTrainData['Survived']==1].std()
meanNotSurvived = normalizedTrainData[normalizedTrainData['Survived']==0].mean()
stdNotSurvived = normalizedTrainData[normalizedTrainData['Survived']==0].std()
```

```python
newTable = pd.DataFrame({'Mean Survived': meanSurvived,
                         'Std Survived': stdSurvived,
                         'Mean Not Survived': meanNotSurvived,
                         'Std Not Survived': stdNotSurvived})
newTable
```

```python
import scipy.stats as stats
import numpy as np 
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(ncols = 1, nrows = len(newTable), figsize = (15,30))

for index, (plotName, row) in enumerate(newTable.iterrows()):
    if index != 0:
        x = np.linspace(-4, 4, 100)
        ax[index].plot(x, stats.norm.pdf(x, row['Mean Survived'], row['Std Survived']), label='Survived', c='blue')
        ax[index].plot(x, stats.norm.pdf(x, row['Mean Not Survived'], row['Std Not Survived']), label='Not Survived', c='red')
        ax[index].legend()
        ax[index].set_title(plotName)
```

 ویژگی ای که برای دادن به ورودی مدل انتخاب میشود، جنسیت خواهد بود زیرا اختلاف دو نمودار زنده ماندن و نماندن از باقی نمودار ها بیشتر است.

# 12.

```python
meanSurv = newTable.loc['Sex']['Mean Survived']
stdSurv = newTable.loc['Sex']['Std Survived']
meanNotSurv = newTable.loc['Sex']['Mean Not Survived']
stdNotSurv = newTable.loc['Sex']['Std Not Survived']
intersectionPoints = np.roots([1 / (2 * (stdSurv ** 2)) - 1 / ( 2 * (stdNotSurv ** 2)), meanNotSurv / (stdNotSurv ** 2) - meanSurv / (stdSurv ** 2), (meanSurv ** 2) / ( 2 * (stdSurv ** 2)) - (meanNotSurv ** 2) / (2 * (stdNotSurv ** 2) - np.log(stdNotSurv / stdSurv))])
intersectionPoints.sort()
def doesSurvived(normalizedTest, intersectionPointsSex):
    if(intersectionPointsSex[0] < normalizedTest < intersectionPointsSex[1]):
        return 0
    return 1
sexCol = testData['Sex']
testData['Sex']= dataLabel.fit_transform(testData['Sex'])
normalizedTestData = (sexCol - sexCol.mean()) / sexCol.std()
finalRes = pd.DataFrame()
finalRes['Survived'] = normalizedTestData.apply(lambda normalizedTest : doesSurvived(normalizedTest, intersectionPoints))
finalRes
```
